```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

file_to_load = "Resources/forestfires.csv"

forestfires2019_df = pd.read_csv(file_to_load)
forestfires2019_df.head()
```

```python
forestfires2019_df.count()
```

```python
months_array = forestfires2019_df['month'].unique()
months_array
```

```python
wind_df =forestfires2019_df.groupby(["month"])["wind"]
avg_wind_df = wind_df.mean()
Average_wind = pd.DataFrame(avg_wind_df)
Average_wind
```

```python
Average_wind_data = "Resources/Average_wind.csv"
Average_wind.to_csv(Average_wind_data,index=False)
```

```python
Average_wind_dict = Average_wind.to_dict('records')
Average_wind_dict
```
