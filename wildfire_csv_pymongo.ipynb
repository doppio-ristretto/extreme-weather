```python
import pymongo as pym
import pandas as pd
import os
import csv
import json


file_to_load = "Resources/MODIS_C6_USA.csv"

USA_wildfires = pd.read_csv(file_to_load)
USA_wildfires.head()
```

```python
USA_wildfires.count()
```

```python
USA_wildfires_data = "Resources/USA_wildfires.csv"
USA_wildfires.to_csv(USA_wildfires_data,index=False)
```

```python
USA_wildfire_dict = USA_wildfires.to_dict('records')
USA_wildfire_dict
```

```python
client = pym.MongoClient('mongodb://localhost:27017')
db = client["USA_wildfires"]
wildfire_table= db["wildfire"]
```

```python
wildfire_table.insert_many(USA_wildfire_dict)
```
