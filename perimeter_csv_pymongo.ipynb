```python
import pymongo as pym
import pandas as pd
import os
import csv
import json


file_to_load = "Resources/Wildfire_Perimeters.csv"

USA_perimeters = pd.read_csv(file_to_load)
USA_perimeters.head()
```

```python
USA_perimeters.count()
```

```python
USA_perimeters_data = "Resources/USA_perimeters.csv"
USA_perimeters.to_csv(USA_perimeters_data,index=False)
```

```python
USA_perimeters_dict = USA_perimeters.to_dict('records')
USA_perimeters_dict
```

```python
client = pym.MongoClient('mongodb://localhost:27017')
db = client["USA_perimeters"]
perimeters_table= db["perimeters"]
```

```python
perimeters_table.insert_many(USA_perimeters_dict)
```
